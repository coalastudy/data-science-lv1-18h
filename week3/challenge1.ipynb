## IsAlone(혼자 탔는가?) 열 추가하기

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
# FamilySize 열을 추가한 후 실습을 이어가겠습니다.
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
# 0이면 동반자와 탑승, 1이면 혼자온것으로 구분
df['IsAlone'] = 1 # IsAlone 열 생성을 목적으로 기본적으로 혼자왔다고 세팅
df.head()
```

```python
# loc 연습: 특정 조건의 정보를 수정하고 싶을 때 주로 사용

# 연습1) Pclass가 1이면 IsAlone이 0이라고 설정 => 1등급 탑승고객은 혼자오지 않았다.
df.loc[ df['Pclass'] == 1, 'IsAlone'] = 0

df
```

```python
# loc 연습: 특정 조건의 정보를 수정하고 싶을 때 주로 사용

# 연습2) Pclass가 1이 아니면 IsAlone이 1이라고 설정 => 2, 3등급(1등급이 아니면) 탑승고객은 혼자왔다.
df.loc[ df['Pclass'] != 1, 'IsAlone'] = 1

df
```

```python
# 미션) FamilySize가 0이면 혼자왔다, 1 이상이면 혼자오지 않았다로 설정
df.loc[ df['FamilySize'] == 0, 'IsAlone'] = 1
df.loc[ df['FamilySize'] > 0, 'IsAlone'] = 0 

df
```
